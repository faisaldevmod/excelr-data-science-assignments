
H0: The average operating cost is equal to the theoretical mean (μ = 4000).  
H1: The average operating cost is different from the theoretical mean (μ ≠ 4000).


```

import math
from scipy.stats import norm

xbar = 3050
theoreticalMu = 4000
popStd = 125
n = 25
alpha = 0.05

z = (xbar - theoreticalMu) / (popStd / math.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

z, p_value
```


Decision Rule:  
If p-value < 0.05, reject the null hypothesis.


```

"Reject H0" if p_value < alpha else "Fail to Reject H0"
```


Conclusion:  
Since the p-value is less than the significance level, we reject the null hypothesis. 
There is sufficient statistical evidence to support the restaurant owners' claim that the 
average operating cost is significantly different from the theoretical cost of 4000.
